```python
# HELPERS TO CALL COMPILED BINARIES
import subprocess, pathlib, importlib
def _bin(name):
    import autoseed
    basedir = pathlib.Path(autoseed.__file__).resolve().parent
    return str(basedir / "bin" / name)
def run_localmax(bg, sig, a, b, c):
    return subprocess.run([_bin("localmax"), bg, sig, str(a), str(b), str(c)], capture_output=True, text=True)
def run_seedext(k, top):
    return subprocess.run([_bin("seedext"), str(k), str(top)], capture_output=True, text=True)
def run_sim():
    return subprocess.run([_bin("sim")], capture_output=True, text=True)
def run_genint(svgp):
    return subprocess.run([_bin("genint"), str(svgp)], capture_output=True, text=True)
print("Binary helpers ready.")
```
